```python
# Stage 1: Hard-coded values

a = -0.5
b = 4.0
c = 20.0
t = 3

temperature = a*t*t + b*t + c

# Writing output to file
file = open("stage1_hardcoded.txt", "w")
file.write("Stage 1: Hard-Coded Values\n")
file.write(f"Time (hours): {t}\n")
file.write(f"Predicted Temperature (°C): {temperature}\n")
file.close()

print("Stage 1 file created: stage1_hardcoded.txt")
```

```python
file = open("stage1_hardcoded.txt", "r")
print(file.read())
file.close()
```

```python
# Stage 2: Keyboard input

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))
t = float(input("Enter time (hours): "))

temperature = a*t*t + b*t + c

# Writing output to file
file = open("stage2_keyboard.txt", "w")
file.write("Stage 2: Keyboard Input\n")
file.write(f"a = {a}, b = {b}, c = {c}, t = {t}\n")
file.write(f"Predicted Temperature (°C): {temperature}\n")
file.close()

print("Stage 2 file created: stage2_keyboard.txt")
```

```python
file = open("stage2_keyboard.txt", "r")
print(file.read())
file.close()
```

```python
file = open("stage3_input.txt", "w")
file.write("-0.4 3.5 22 5")
file.close()

print("Input file created: stage3_input.txt")
```

```python
file = open("stage3_input.txt", "r")
a, b, c, t = map(float, file.read().split())
file.close()

temperature = a*t*t + b*t + c

file = open("stage3_output.txt", "w")
file.write("Stage 3: File Input (Single Set)\n")
file.write(f"Time: {t}\n")
file.write(f"Predicted Temperature (°C): {temperature}\n")
file.close()

print("Output file created: stage3_output.txt")
```

```python
file = open("stage3_output.txt", "r")
print(file.read())
file.close()
```

```python
file = open("stage4_input.txt", "w")
file.write("-0.5 4 20 3\n")
file.write("-0.3 3 18 4\n")
file.write("-0.2 2.5 25 6")
file.close()

print("Input file created: stage4_input.txt")
```

```python
file = open("stage4_input.txt", "r")
output = open("stage4_output.txt", "w")

output.write("Stage 4: File Input (Multiple Sets)\n\n")

for line in file:
    a, b, c, t = map(float, line.split())
    temperature = a*t*t + b*t + c
    output.write(f"Time: {t} hrs -> Temperature: {temperature} °C\n")

file.close()
output.close()

print("Output file created: stage4_output.txt")
```

```python
file = open("stage4_output.txt", "r")
print(file.read())
file.close()
```
